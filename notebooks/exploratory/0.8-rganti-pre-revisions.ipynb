```python
def define_n_initial(num_bins):
    n_initial = np.zeros(num_bins, dtype=float)

    n_b_cells = 25
    index = (num_bins + 1) / 2

    mu, sigma = 3., 1.

    s = np.random.lognormal(mu, sigma, n_b_cells)
    count, bins = np.histogram(s, bins=range(0, 400, 400/index))

    n_initial[:len(count)] = count

    s = np.random.lognormal(mu, sigma, n_b_cells)
    count, bins = np.histogram(s, bins=range(0, 400, 400/index))

    n_initial[-len(count):] = count[::-1]

    return n_initial
```
