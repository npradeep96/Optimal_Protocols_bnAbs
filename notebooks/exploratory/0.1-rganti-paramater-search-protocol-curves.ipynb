```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
%config InlineBackend.figure_format = 'svg'
import sys
import pickle
import numpy as np
np.set_printoptions(suppress=True)
import matplotlib.pyplot as plt
import seaborn as sns
from src.visualization.visualize_fitness import InjectionKlDistance, ProcessProtocolData
```

```python
sigma_1_range = np.sort(np.logspace(1.0, -1.0, num=15))[::-1]
```

```python
path = "../../data/raw/check_protocol_curves/"
protocol_data = ProcessProtocolData(path)
protocol_data.load_arrays()
```

```python
sns.set_palette(sns.color_palette("hls", len(protocol_data.kl1_array)))
for i in range(len(protocol_data.kl1_array)):
    plt.errorbar(protocol_data.kl2_matrix[i], protocol_data.mean_bnab_array[i], marker='o', 
                 label='$D(p_{0} || f_{1}) = ' + '%.2f' % protocol_data.kl1_array[i] + ', \\sigma_{1} = %.2f' % protocol_data.sigma_1_range[i] + ' $')    

plt.legend(bbox_to_anchor=(1.05, 1.00))
plt.xlabel("$D(p_{1} || f_{2})$")
plt.ylabel("Number of bnAbs/GC")
plt.title("Prime and Boost Protocols")
```
